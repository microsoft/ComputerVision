```python
import sys
sys.path.append("../../")

import matplotlib.pyplot as plt
import numpy as np
from utils_ic.plot_utils import plot_roc_curve, plot_precision_recall_curve, ResultsWidget
```

```python
# Binary-class classification testcase
BINARY_Y_TRUE = [0, 0, 1, 1]
BINARY_Y_SCORE = [0.1, 0.4, 0.35, 0.8]
BINARY_CLASSES = [0, 1]

# Multi-class classification testcase
MULTI_Y_TRUE = [0, 0, 1, 1, 2, 2]
MULTI_Y_SCORE = [
    [0.1, 0.9, 0.0], 
    [0.4, 0.2, 0.4],
    [0.35, 0.15, 0.5],
    [0.1, 0.8, 0.1],
    [0.2, 0.5, 0.3],
    [0.0, 0.1, 0.9],
]
MULTI_CLASSES = [0, 1, 2]
```

```python
# Binary-class plot
y_true = np.array(BINARY_Y_TRUE)
y_score = np.array(BINARY_Y_SCORE)

plt.subplots(2, 2)

plt.subplot(1, 2, 1)
plot_roc_curve(y_true, y_score, BINARY_CLASSES, False)
plt.subplot(1, 2, 2)
plot_precision_recall_curve(y_true, y_score, BINARY_CLASSES, False)
plt.show()
```

```python
# Multi-class plot
y_true = np.array(MULTI_Y_TRUE)
y_score = np.array(MULTI_Y_SCORE)

plt.subplots(2, 2)

plt.subplot(1, 2, 1)
plot_roc_curve(y_true, y_score, MULTI_CLASSES, False)
plt.subplot(1, 2, 2)
plot_precision_recall_curve(y_true, y_score, MULTI_CLASSES, False)
plt.show()
```
